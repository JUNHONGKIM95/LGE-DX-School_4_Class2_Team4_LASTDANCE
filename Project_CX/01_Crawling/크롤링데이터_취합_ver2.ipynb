```python
import pandas as pd
import os
```

```python
today_date = '251212'
ver = 3

fd_path_1 = f'./results_취합/{today_date}/요리 전/'
fd_path_2 = f'./results_취합/{today_date}/요리 중/'
fd_path_3 = f'./results_취합/{today_date}/요리 후/'
```

```python
fileEx = '.xlsx'
```

### 크롤링 데이터 취합

- url 제외

```python
### 요리 전

xlsx_list = [file for file in os.listdir(fd_path_1) if file.endswith(fileEx)]
print("엑셀 파일 총 개수: ", len(xlsx_list))

n1 = 0
df1 = pd.DataFrame()
for xl in xlsx_list:
    df = pd.read_excel(fd_path_1+xl)
    
    # 네이버 카페 -> 날짜 형식 통일
    if xl.split('_')[0] == '네이버' and xl.split('_')[1] == '카페':
        df['Date'] = df['Date'].str.replace('-', '')
        df['src'] = '네이버 카페'
        
    # 네이버 블로그
    elif xl.split('_')[0] == '네이버' and xl.split('_')[1] == '블로그':
        df['src'] = '네이버 블로그'
        
    # 블라인드
    elif xl.split('_')[0] == 'blind':
        df['src'] = '블라인드'
        
    # 82cook
    elif xl.split('_')[0] == '82cook':
        df['src'] = '82cook'

    df['Date'] = df['Date'].apply(lambda x: str(x))
    
    df1 = pd.concat([df1, df])
    print(f'{xl}: {len(df)}')    


df1['Contents'] = df1['Contents'].str.replace(r'http\S+','[URL]', regex=True) # url 제거
n1 = len(df1)

print("[요리 전] 총합: ", n1)
```

```python
### 요리 중

xlsx_list = [file for file in os.listdir(fd_path_2) if file.endswith(fileEx)]
print("엑셀 파일 총 개수: ", len(xlsx_list))

n2 = 0
df2 = pd.DataFrame()
for xl in xlsx_list:
    df = pd.read_excel(fd_path_2+xl)
    
        # 네이버 카페 -> 날짜 형식 통일
    if xl.split('_')[0] == '네이버' and xl.split('_')[1] == '카페':
        df['Date'] = df['Date'].str.replace('-', '')
        df['src'] = '네이버 카페'
        
    # 네이버 블로그
    elif xl.split('_')[0] == '네이버' and xl.split('_')[1] == '블로그':
        df['src'] = '네이버 블로그'
        
    # 블라인드
    elif xl.split('_')[0] == 'blind':
        df['src'] = '블라인드'
        
    # 82cook
    elif xl.split('_')[0] == '82cook':
        df['src'] = '82cook'

    df['Date'] = df['Date'].apply(lambda x: str(x))
            
    df2 = pd.concat([df2, df])
    print(f'{xl}: {len(df)}')
    
df2['Contents'] = df2['Contents'].str.replace(r'http\S+','[URL]', regex=True) # url 제거
n2 = len(df2)

print("[요리 중] 총합: ", n2)
```

```python
### 요리 후

xlsx_list = [file for file in os.listdir(fd_path_3) if file.endswith(fileEx)]
print("엑셀 파일 총 개수: ", len(xlsx_list))

n3 = 0
df3 = pd.DataFrame()
for xl in xlsx_list:
    df = pd.read_excel(fd_path_3+xl)
    
    # 네이버 카페 -> 날짜 형식 통일
    if xl.split('_')[0] == '네이버' and xl.split('_')[1] == '카페':
        df['Date'] = df['Date'].str.replace('-', '')
        df['src'] = '네이버 카페'
        
    # 네이버 블로그
    elif xl.split('_')[0] == '네이버' and xl.split('_')[1] == '블로그':
        df['src'] = '네이버 블로그'
        
    # 블라인드
    elif xl.split('_')[0] == 'blind':
        df['src'] = '블라인드'
        
    # 82cook
    elif xl.split('_')[0] == '82cook':
        df['src'] = '82cook'

    df['Date'] = df['Date'].apply(lambda x: str(x))
    
    df3 = pd.concat([df3, df])
    print(f'{xl}: {len(df)}')
    
df3['Contents'] = df3['Contents'].str.replace(r'http\S+','[URL]', regex=True) # url 제거
n3 = len(df3)

print("[요리 후] 총합: ", n3)
```

### 데이터 전처리
    1. 결측치 제거
    2. 불용어 처리
    3. 중복 제거

```python
# 1. 결측치 제거
print('[요리 전] 결측치 제거 전: ', len(df1))
# df1 = df1.dropna()
df1 = df1.dropna(subset=['Contents'])
print('[요리 전] 결측치 제거 후: ', len(df1))


print('[요리 중] 결측치 제거 전: ', len(df2))
# df2 = df2.dropna()
df2 = df2.dropna(subset=['Contents'])
print('[요리 중] 결측치 제거 후: ', len(df2))


print('[요리 후] 결측치 제거 전: ', len(df3))
# df3 = df3.dropna()
df3 = df3.dropna(subset=['Contents'])
print('[요리 후] 결측치 제거 후: ', len(df3))

print("*결측치 제거 후 총합: ", len(df1)+len(df2)+len(df3))
```

```python
# 2. 불용어 처리
stopwords = ['광고', '알바', 
             '비밀번호', '회원', '계정', '로그인', '정지', 
             '자격', '규정', '공지', 
             '댓글', '게시', '삭제', '운영', 
             '사이트', '커뮤니티', '정치',
             '모집','판 · 매 · 합 · 니 · 다 l 그 · 냥 · 드 · 려 · 요','협찬을','홍보','렌탈마트','개월 무료', # 251211 추가
             '황금점', '이사스토리' , '내손안에정육점', 'LG전자 구독 좋아요 이벤트 참여', '공동구매가 진행중입니다', '강동구 ㅊ병원' # 251212 추가
             ]

#########################################
print("[요리 전] Before:", len(df1))

# stopwords 중 하나라도 포함하면 True
mask = df1['Contents'].str.contains('|'.join(stopwords), na=False)

# 해당 행 제거
df1 = df1[~mask].reset_index(drop=True)

print("[요리 전] After:", len(df1))
#########################################

#########################################
print("[요리 중] Before:", len(df2))

# stopwords 중 하나라도 포함하면 True
mask = df2['Contents'].str.contains('|'.join(stopwords), na=False)

# 해당 행 제거
df2 = df2[~mask].reset_index(drop=True)

print("[요리 중] After:", len(df2))
#########################################

#########################################
print("[요리 후] Before:", len(df3))

# stopwords 중 하나라도 포함하면 True
mask = df3['Contents'].str.contains('|'.join(stopwords), na=False)

# 해당 행 제거
df3 = df3[~mask].reset_index(drop=True)

print("[요리 후] After:", len(df3))
#########################################

print("="*15)
print("[요리 전]: ", len(df1))
print("[요리 중]: ", len(df2))
print("[요리 후]: ", len(df3))
print("*불용어 처리 후 총합: ", len(df1)+len(df2)+len(df3))
```

```python
# 3. 중복 제거

df_total = pd.DataFrame()
df_total = pd.concat([df_total, df1])
df_total = pd.concat([df_total, df2])
df_total = pd.concat([df_total, df3])

df_total = df_total.drop_duplicates()

print("*중복 제거 후 총합", len(df_total))
```

### 수집 연도 기준 컬럼 추가

```python
df_total['Date_YYYY'] = df_total['Date'].apply(lambda x: str(x)[:4]) # YYYY
df_total.loc[df_total['Date_YYYY'].str.contains(':'), 'Date_YYYY'] = '2025' # 오늘 날짜인 경우 시간만 추출됨 (ex. 12:55)
df_total.head()
```

```python
df_total['Date_YYYY'].value_counts()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# value_counts 결과
vc = df_total['Date_YYYY'].value_counts()

# index를 문자열로 변환
vc.index = vc.index.astype(str)

# nan을 맨 앞에 두고 나머지는 정렬
years = [y for y in vc.index if y != 'nan']
years_sorted = sorted(years, key=lambda x: int(x))
years_sorted = ['nan'] + years_sorted

# 정렬 적용
vc_sorted = vc.reindex(years_sorted)

# 시각화
pastel_red = "#FF9999" # 파스텔 레드 계열 색상 (연한 핑크빛)
fig, ax = plt.subplots(figsize=(14, 8))
bars = ax.bar(vc_sorted.index, vc_sorted.values, color=pastel_red)

# 막대 위에 y값 표시
ax.bar_label(bars, padding=3)

ax.set_xlabel("Year", fontsize=15)
ax.set_ylabel("Count", fontsize=15)
ax.set_title("Crawling Data Overview (2021~2025)", fontsize=20)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# value_counts 결과
vc = df_total_recent5yr['Date_YYYY'].value_counts()

# index를 문자열로 변환
vc.index = vc.index.astype(str)

# nan을 맨 앞에 두고 나머지는 정렬
years = [y for y in vc.index if y != 'nan']
years_sorted = sorted(years, key=lambda x: int(x))
years_sorted = ['nan'] + years_sorted

# 정렬 적용
vc_sorted = vc.reindex(years_sorted)

# 시각화
pastel_red = "#FF9999" # 파스텔 레드 계열 색상 (연한 핑크빛)
fig, ax = plt.subplots(figsize=(14, 8))
bars = ax.bar(vc_sorted.index, vc_sorted.values, color=pastel_red)

# 막대 위에 y값 표시
ax.bar_label(bars, padding=3, fontsize=18)

ax.set_xlabel("Year", fontsize=15)
ax.set_ylabel("Count", fontsize=15)
ax.set_title("Crawling Data Overview (2021~2025)", fontsize=20)
# plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 한글 폰트 설정 (윈도우)
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

src_counts = df_total_recent5yr['src'].value_counts()

# 비율 + 건수 표시 함수
def autopct_format(pct):
    total = src_counts.sum()
    count = int(round(pct * total / 100))
    return f'{count}건\n({pct:.1f}%)'

plt.figure(figsize=(10, 8))

# labels=None 으로 설정하여 파이 조각에는 글자를 넣지 않음
wedges, texts, autotexts = plt.pie(
    src_counts,
    labels=None,
    autopct=autopct_format,
    startangle=90,
    colors=plt.cm.Pastel1.colors
)

# autopct 텍스트 폰트 크기 조절
for autotext in autotexts:
    autotext.set_fontsize(12)

# 📌 범례 추가 (src 이름 표시)
plt.legend(
    wedges,                      # 조각(=색깔)과 매칭
    src_counts.index,            # src 종류
    title="Source",              # 범례 제목
    loc="center left",
    bbox_to_anchor=(1.05, 0.5),  # 차트 오른쪽 바깥에 배치
    fontsize=15                  # 범례 텍스트 폰트 크기
)

# 제목
plt.title("Source Distribution (2021–2025)", fontsize=16, fontweight="bold")

plt.axis('equal')
plt.tight_layout()
plt.show()
```
