### 1. 냉장고

```python
review_text_list = []
review_score_list = []

for idx, url in enumerate(url_lst_refrig):
    print('-'*15)
    print("url #: ", idx+1)
    
    chrome_options = Options()
    chrome_options.add_experimental_option("detach", True)
    chrome_options.add_experimental_option("excludeSwitches", ["enable-logging"])
    # chrome_options.add_argument("headless") # headless option
    driver = wb.Chrome(options=chrome_options)
    driver.get(url)
    time.sleep(1)
    
    # '리뷰' 버튼 클릭
    review_btn = driver.find_element(By.ID, 'tabMenuDetail2').click()
    time.sleep(1)
    
    
    # 스크롤 내리기 (전체화면(body)에 END 키를 전송)
    time.sleep(1)
    
    # iframe
    driver.switch_to.frame('ifrmReview')
    time.sleep(1)
    
    # '리뷰 더보기' 버튼 클릭
    while True:
        try:
            driver.find_element(By.CSS_SELECTOR, '#review-list-page-area > div > button').click()
            time.sleep(1)

        except:
            print('리뷰 더보기 클릭 완료')
            break
    
    
    # 리뷰 텍스트 가져오기
    review_texts = driver.find_elements(By.CSS_SELECTOR, 'div.c_product_review_cont > div.cont')
    review_scores = driver.find_elements(By.CSS_SELECTOR, 'div.c_product_review_cont > p.grade em')


    for i in tqdm(range(len(review_texts))):
        try:
            review_text = review_texts[i].find_element(By.CSS_SELECTOR, 'div.cont_text_wrap').text
            review_score = review_scores[i].text
            
            review_text_list.append(review_text)
            review_score_list.append(review_score)
        except:
            continue

    print(len(review_text_list), len(review_score_list))
    print('-'*15)
    
    driver.quit()
    

# dataframe 저장
review_df = pd.DataFrame([review_text_list, review_score_list]).T
review_df.columns = ['review', 'score']

review_df.to_excel('./results/11street_refrigerator.xlsx', index=False)
```
